```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.naive_bayes import GaussianNB, CategoricalNB
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.model_selection import train_test_split
```

```python
DATA_LOADER = pd.read_csv('winequality-white.csv', sep=";")
DATA = DATA_LOADER.iloc[:, :-1]
DATA_LABELS = DATA_LOADER.iloc[:, -1]
# print(DATA_LABELS)

X_train, X_test, y_train, y_test = train_test_split(DATA, DATA_LABELS, train_size=0.8, test_size=0.2)
bayes = GaussianNB()
bayes.fit(X_train, y_train)
```

```python
predicted = bayes.predict(X_test)
print("Accuracy: {}%".format(round(accuracy_score(y_test, predicted) * 100)))
```

```python
cm = confusion_matrix(y_test, predicted)
sns.heatmap(cm, annot=True, fmt='d')
plt.show()
```
