```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
from sklearn.naive_bayes import CategoricalNB, GaussianNB
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
DATA_LOADER = pd.read_csv('iris_data.csv')
DATA_TRAIN = DATA_LOADER[['sepalLength', 'sepalWidth', 'petalLength', 'petalWidth']]
DATA_LABELS = DATA_LOADER[['nhan']]
X_train, X_test, y_train, y_test = train_test_split(DATA_TRAIN, DATA_LABELS, train_size=0.8, test_size=0.2)
#print(len(X_test))
bayes = GaussianNB()
bayes.fit(X_train, y_train)
```

```python
predicted = bayes.predict(X_test)
print("Accuracy: {}%".format(round(accuracy_score(y_true=y_test, y_pred=predicted) * 100)))
cm = confusion_matrix(y_true=y_test, y_pred=predicted)
sns.heatmap(cm, annot=True, fmt='d')
plt.xlabel(np.array(['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']))
plt.ylabel(np.array(['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']))
plt.show()
```
